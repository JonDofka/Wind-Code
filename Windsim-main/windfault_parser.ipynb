```python
import pandas as pd # type: ignore
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

Parsing the csvs:

1. Remove first two lines. 
2. Separate files with a FRT Monitor and those without. 


```python
# Directory containing the CSV files
directory = 'Wind_Faults'

# Initialize an empty list to store filenames with data in the 21st column
frt_files = []
ug_a = {}
# Loop through all files in the directory
for filename in os.listdir(directory):
    if 'frt' in filename:
        frt_files.append(filename)

# Read the CSV files and extract the 'Ug_a' column
for file in frt_files:
    file_path = os.path.join(directory, file)
    frt = pd.read_csv(file_path)
    if 'Ug_a' in frt.columns:
        ug_a[file] = frt['Ug_a']

# Plotting the 'Ug_a' columns from all files
plt.figure(figsize=(10, 6))
for file, data in ug_a.items():
    plt.plot(data, label=file)

plt.xlabel('Index')
plt.ylabel('Ug_a')
plt.title('Comparison of Ug_a across different files')
plt.legend()
plt.show()
```
